- get the length_inches from the title
- get the gender from the title, if present
- the title has information that is not in the category, so get all words other than 'Imported' and 'Koi'

```python
# create a function to check if the first word is a number, if so, return it in new feature length_inches

def get_length(string):
  if string.split()[0][0].isdigit():
    return string.split()[0]
```

```python
# apply the function to the whole dataset

nextdaykoi['length_inches'] = nextdaykoi['title'].apply(get_length)
nextdaykoi['length_inches']
```

```python
# check how many are nulls

nextdaykoi['length_inches'].isna().sum()
```

```python
# extract the gender if present

for word in nextdaykoi['title'][20].split():
  if word.lower() in ['male', 'female']:
    print(word)
```

```python
# create a function to extract the gender into a new column

def get_gender(string):
  for word in string.split():
    if word.lower() in ['male', 'female']:
      return word
```

```python
# apply the function to the dataset

nextdaykoi['gender'] = nextdaykoi['title'].apply(get_gender)
nextdaykoi['gender']
```

```python
# check value_counts

nextdaykoi['gender'].value_counts()
```

- there is not much data, this feature may need to be dropped later

```python
nextdaykoi['title'][100]
```

```python
# extract other words as tags (words that are not gender-related, "imported", "lot", "of" or "koi" and starts with a alphabet letter)

words_to_ignore = ['imported', 'lot', 'of', 'koi', 'assorted', 'and']

tags_list = []
for word in nextdaykoi['title'][100].split():
  if word.lower() not in words_to_ignore and word[0].isalpha():
    tags_list.append(word)

tags_list
```

```python
# create a function to extract the tags

def get_tags_from_title(string):
  words_to_ignore = ['imported', 'lot', 'of', 'koi', 'assorted', 'and']
  tags_list = []
  for word in string.split():
    if word.lower() not in words_to_ignore and word[0].isalpha():
      tags_list.append(word)
  return tags_list
```

```python
# apply the function to the dataset

nextdaykoi['tags_from_title'] = nextdaykoi['title'].apply(get_tags_from_title)
nextdaykoi['tags_from_title']
```

```python
# join the category feature with the tags_from_title to create the final tags feature
# check the type of the data

type(nextdaykoi['category'][0])
```

```python
# the category is a string that is a list of strings, turn it back into a list of strings

nextdaykoi['category'][0]
```

```python
# the category is a string that is a list of strings, turn it back into a list of strings
# strip the double quotes

print(nextdaykoi['category'][0].strip('"'))
```

```python
# strip the brackets

print(nextdaykoi['category'][0].strip('"').strip('[]'))
```

```python
# split by ", "

print(nextdaykoi['category'][0].strip('"').strip('[]').split(", "))
```

```python
# strip the single quotes

print([item.strip("'") for item in nextdaykoi['category'][0].strip('"').strip('[]').split(", ")])
```

```python
# create a function to turn the string of list of strings into a list of strings

def string_to_list(string):
  return [item.strip("'") for item in string.strip('"').strip('[]').split(", ")]
```

```python
# apply the function to the dataset

nextdaykoi['tags_from_category'] = nextdaykoi['category'].apply(string_to_list)
nextdaykoi['tags_from_category']
```

```python
# double check the type

type(nextdaykoi['tags_from_category'][0])
```

```python
# combine the two lists of tags into a final tags column

nextdaykoi['tags'] = nextdaykoi['tags_from_title'] = nextdaykoi['tags_from_category']
nextdaykoi['tags']
```

# Data from Sacramentokoi - 423 pictures

```python
# check the info

sacramentokoi.info()
```

```python
# check the data

sacramentokoi.head()
```

```python
# check the category types

sacramentokoi['category'].value_counts()
```

- there is Female label but no Male label

```python
# check more rows of the title

sacramentokoi['title'].head(30)
```

```python
# get the words in the title as tags

tags_list = []

for word in sacramentokoi['title'][0].split():
  if word[0].isalpha():
    tags_list.append(word)
  else:
    break

tags_list
```

```python
# create a function to get the list of tag words

def get_tags(string):
  tags_list = []
  for word in string.split():
    if word[0].isalpha():
      tags_list.append(word)
    else:
      return tags_list
```

```python
# apply the function to the entire dataset

sacramentokoi['tags_from_title'] = sacramentokoi['title'].apply(get_tags)
sacramentokoi['tags_from_title']
```

```python
# check the value counts

sacramentokoi['tags_from_title'].value_counts()
```

```python
# check if the category is already included in tag

sacramentokoi[['category', 'tags_from_title']]
```

```python
# for the category, if it is not gender-related, or the word 'Koi', and if not already in list of tags, then add to list of tags

sacramentokoi['tags_from_category'] = None

for index, row in sacramentokoi.iterrows():
  words_to_skip = ['koi', 'female', 'male']
  words_list = []

  # in case tags is empty, fallback to empty list
  tags = [tag.lower() for tag in row['tags_from_title']] if isinstance(row['tags_from_title'], list) else []

  for word in row['category'].split():
    if word.lower() not in tags and word.lower() not in words_to_skip:
      words_list.append(word)
  sacramentokoi.at[index, 'tags_from_category'] = words_list
```

```python
# check how many unique tags acquired from the category feature

sacramentokoi['tags_from_category'].value_counts()
```

```python
# combine the tags_from_title to tags_from_category into one final tags feature

sacramentokoi['tags'] = sacramentokoi['tags_from_title'] + sacramentokoi['tags_from_category']
sacramentokoi['tags']
```

```python
# check for nulls

sacramentokoi['tags'].isna().sum()
```

```python
# get the gender

for word in sacramentokoi['category'][1].split():
  if word.lower() in ['female', 'male']:
    print(word)
```

```python
# create a function to get the gender

def get_gender_sacramento(string):
  for word in string.split():
    if word.lower() in ['female', 'male']:
      return word
```

```python
# apply the function to the dataset and create new gender feature

sacramentokoi['gender'] = sacramentokoi['category'].apply(get_gender_sacramento)
sacramentokoi['gender']
```

```python
# check the counts

sacramentokoi['gender'].value_counts()
```

```python
# get fish size, length_inches

for string in sacramentokoi['title'][0].split():
  if string[0].isdigit():
    print(string)
    break
```

```python
# create a function to extract the fish size

def get_size(string):
  for word in string.split():
    if word[0].isdigit():
      return word
```

```python
# apply the function to the whole dataset

sacramentokoi['length_inches'] = sacramentokoi['title'].apply(get_size)
sacramentokoi['length_inches']
```

```python
# rename column "Unnamed: 0" to "uuid"

sacramentokoi = sacramentokoi.rename(columns={'Unnamed: 0': 'uuid'})
```
