# Data from Grandkoi - 367 pictures

```python
# check the info

grandkoi.info()
```

```python
# check the data

grandkoi.head()
```

```python
# check the details

grandkoi['details']
```

```python
# extract the breeder name

grandkoi['breeder'] = grandkoi['details'].str.extract(r'Breeder:\s*(\w+)')
grandkoi['breeder']
```

```python
# extract the length in inches

grandkoi['length_inches'] = grandkoi['details'].str.extract(r'Inches:\s*(\d+)').astype(float)
grandkoi['length_inches']
```

```python
# check the updated dataframe

grandkoi
```

```python
# remove "high quality" and "sold" from the title into individual words

grandkoi['title'][0].lower().replace('high quality', '')
```

```python
# remove "high quality" and "sold" from the title into individual words

words_to_remove = ['high', 'quality', 'sold']

def remove_words(string):
  return [word.strip('-') for word in string.split() if word.lower() not in words_to_remove]
```

```python
# test the function on 1 row

print(f'before: {grandkoi["title"][0]}')
print(f'after: {remove_words(grandkoi["title"][0])}')
```

```python
# apply the function on all rows

grandkoi['tags'] = grandkoi['title'].apply(remove_words)
grandkoi['tags']
```

```python
# rename the Unnamed: 0 column to uuid

grandkoi = grandkoi.rename(columns={'Unnamed: 0': 'uuid'})
grandkoi
```

```python
# export the cleaned dataframe to csv

# grandkoi.to_csv('grandkoi_df.csv', index=False)
```

```python
# export the dataframe to json to maintain list formats

with open('grandkoi.json', 'w') as f:
  json.dump(grandkoi.to_dict(orient='records'), f)
```

# Data from Kloubec - 517 pictures

```python
# check the info

kloubec.info()
```

```python
# check the head

kloubec.head()
```

```python
# check the description

kloubec['description'].value_counts()
```

- only 8 out of 517 observations had a unique description, i will drop this feature


```python
# check the out_of_stock

kloubec['out_of_stock'].value_counts()
```

- all of the observations are False, so this feature is meaningless, I will drop it

```python
# drop description and out_of_stock features

kloubec = kloubec.drop(columns=['description', 'out_of_stock'])
```

```python
# check the difference between category and the title for the first 5 rows

for index, row in kloubec.head(5).iterrows():
  print(index)
  print(row['title'])
  print(row['category'])
```

- category is not useful, only the title contains relevant information to the observation. I will drop the category feature as well.

```python
# drop category feature

kloubec = kloubec.drop(columns=['category'])
kloubec
```

```python
# check the title feature

for _, row in kloubec.head(10).iterrows():
  print(row['title'])
```

From the title, I can extract:
- length_inches
- tags, all words except "Koi" and in words in double quotes

```python
sample_row = kloubec['title'][0]
sample_row
```

```python
# first check the words in double quotes to see if they are useful

import re

re.search(r'\s"([^"]+)"', sample_row)
```

```python
# apply it on whole dataset

kloubec['in_quotes'] = kloubec['title'].str.extract(r'\s"([^"]+)"')
kloubec['in_quotes'].value_counts()
```

- these terms are not related to koi variety, and are therefore irrelevant. the column can be dropped, and the terms inside of double quotes in the title can be safely ignored

```python
# drop in_quotes feature

kloubec = kloubec.drop(columns=['in_quotes'])
kloubec.head()
```

```python
# create a function to extract length_inches and tags from the title feature

def get_length(string):
  for word in string.split():
    if word[0].isdigit():
      return word
```

```python
# apply the function to the entire dataset

kloubec['length_inches'] = kloubec['title'].apply(get_length)
kloubec['length_inches'].value_counts()
```

```python
# create a function to extract the tags from the title

def get_tags(string):
  list_of_tags = []
  for word in string.split():
    if word[0].isalpha() and word.lower() != 'koi':
      list_of_tags.append(word)
  return list_of_tags
```

```python
# apply the function to the entire dataset

kloubec['tags'] = kloubec['title'].apply(get_tags)
kloubec['tags'].head(20)
```

```python
# check the dataframe

kloubec
```

```python
# rename column 'Unnamed: 0' to 'uuid'

kloubec = kloubec.rename(columns={'Unnamed: 0': 'uuid'})
kloubec.head()
```

```python
# export the dataframe to csv

# kloubec.to_csv('kloubec_df.csv', index=False)
```

```python
# export the dataframe to json to maintain list formats

with open('kloubec.json', 'w') as f:
  json.dump(kloubec.to_dict(orient='records'), f)
```
