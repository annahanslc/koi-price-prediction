```python
data_collected = {}

def scrape_page(page_num):
    url  = START_URL.format(page_num)
    resp = requests.get(url)
    if resp.status_code != 200:
        return False    # no such page

    soup = BeautifulSoup(resp.text, "html.parser")

    products = soup.select("li.product")
    if not products:
        return False    # no products → end of pagination

    for prod in products:
        card = prod.select_one("article.card.epic-product-card")

        # 1) ID, name & price are stored as data‑attributes on the <article>
        sku           = card["data-entity-id"]
        title         = card["data-name"]           .strip()
        price         = card["data-product-price"]  .strip()
        category      = [
            c.strip() for c in card["data-product-category"].split(",")
            if c.strip()
        ]

        # 2) product page link
        link = card.select_one("a.card-figure__link")["href"]

        # 3) main image URL
        img_tag = card.select_one("figure.card-figure img.card-image")
        # BigCommerce often lazy‑loads, but the <img> you showed has a real src
        img_url = img_tag.get("src") or img_tag.get("data-src")
        img_url = urljoin(BASE_URL, img_url)

        # 4) short description (first <p> under .card-description)
        desc_tag = card.select_one("div.card-description p")
        description = desc_tag.get_text(strip=True) if desc_tag else ""

        # 5. out‑of‑stock?
        out_of_stock = bool(prod.select_one("div.out-of-stock-label"))


        # 6) download & save image
        ext      = os.path.splitext(img_url)[1].split("?")[0] or ".jpg"

        # generate a short 6‑hex‑digit ID
        unit_id = uuid.uuid4().hex[:6]  # e.g. '9f1c2a'

        # remove illegal filename chars
        safe_title = re.sub(r'[\\/*?:"<>|]', "", title)
        # replace spaces (and dashes) with underscores
        safe_title = safe_title.replace(" ", "_").replace("–", "-")

        filename = f"{unit_id}_{safe_title}{ext}"
        filepath     = os.path.join(OUTPUT_DIR, filename)

        # download in streaming mode
        resp = requests.get(img_url, stream=True)
        resp.raise_for_status()
        with open(filepath, "wb") as f:
            for chunk in resp.iter_content(chunk_size=8192):
                if not chunk:
                    continue
                f.write(chunk)


        # 7. output or store your data
        data_collected[unit_id] = {
            "link":     link,
            "title":    title,
            "description": description,
            "category":  category,
            "out_of_stock":  out_of_stock,
            "price":    price,
            "sku":      sku,
            "image":    filepath,      # path where you saved it
            "image_url": img_url,      # original URL
        }
        pass
    return True
```

```python
page = 1
while True:
    print(f"Scraping page {page}…")
    success = scrape_page(page)
    if not success:
        print("No more pages. Done.")
        break
    page += 1
    time.sleep(1)  # be polite
```

```python
df = pd.DataFrame(data_collected)
```
