# EDA - 1 - Checking null and dropping features and observations

```python
# imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import numbers
```

```python
# import merged data

with open('data/merged_df.json', 'r') as f:
  data = json.load(f)

df = pd.DataFrame(data)
```

```python
# check the merged dataset

df.head()
```

```python
# check the info

df.info()
```

```python
# create a dataframe for the info

# Get the data types
dtypes = df.dtypes

# Get the non-null counts
counts = df.count()

# Create a new DataFrame
info_df = pd.DataFrame({'Column': dtypes.index, 'Dtype': dtypes.values, 'Non-Null Count': counts.values})

info_df
```

```python
# calculate the total number of observations

total_observations = info_df[info_df['Column'] == 'uuid']['Non-Null Count'].values[0]
total_observations
```

```python
# determine the features that have more than 10% nulls

cols_to_drop = []

for index, row in info_df.iterrows():
  if row['Non-Null Count'] / total_observations < 0.9:
    cols_to_drop.append(row['Column'])

cols_to_drop
```

```python
# drop the features that have more than 10% nulls

df = df.drop(columns=cols_to_drop)
df
```

```python
# other columns to drop (irrelevant): image_url, link, sku, source, title

df = df.drop(columns=['image_url', 'link', 'sku', 'source', 'title'])
df.head()
```

```python
# drop all observations that is True in multiple_fish

df = df[df['multiple_fish']==False]
df
```

```python
# now drop the multiple_fish feature

df = df.drop(columns=['multiple_fish'])
df
```

```python
# check for any more ranges in length_inches

df[df['length_inches'].str.contains('-', na=False)]
```

```python
# drop these 3 observations that have ranges in the length

df = df[~df['length_inches'].str.contains('-', na=False)]
df
```

```python
# check the datatypes

df.info()
```

```python
# change length_inches and price to integers
# create a function

def clean_length(string):
  if isinstance(string, numbers.Number):
    return float(string)
  if isinstance(string, str) and string.strip():
    if string[0].isdigit():
      return float(string.strip("”").strip('″').strip('"'))
  else:
    return np.nan
```

```python
# apply the function

df['length'] = df['length_inches'].apply(clean_length)
df
```

```python
# drop length_inches

df = df.drop(columns='length_inches')

df.info()
```

```python
# check for any non-numeric characters in the price feature

df['price'].value_counts()
```

```python
# check for abnormally long price, which may indicate this is extra information

df['price_length'] = df['price'].astype(str).str.len()
df
```

```python
# create a function to clean the price string

def price_to_float(string):
  price_list = []

  if isinstance(string, numbers.Number):
    return string

  if isinstance(string, str) and string.split():
    for word in string.split('$'):
      if word.split() and word[0].isdigit():
        word = word.replace(r'[a-zA-Z]','')
        word = word.replace(',','')
        word = word.strip('.')
        price_list.append(word)

    if price_list:
      return float(min(price_list))

    else:
      print(string)
```

```python
# apply the function

df['cleaned_price'] = df['price'].apply(price_to_float)
df
```

```python
# check the nulls

df[df['cleaned_price'].isna()]
```

- the nulls indeed don't have prices, they will eventually be dropped

```python
# drop the original price column and rename 'cleaned_price' to 'price'ArithmeticError

df = df.drop(columns='price')
df = df.rename(columns={'clean_price': 'price'})

df.head()
```

```python
# check the null count

df.info()
```

```python
# drop all observations that have nulls

df = df.dropna()
df.info()
```
